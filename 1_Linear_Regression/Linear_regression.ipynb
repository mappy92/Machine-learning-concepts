```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)', fontsize=20)
plt.ylabel('price (US$)', fontsize=20)
plt.scatter(df.Area.values, df.Prices.values, color='green', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['Area']], df['Prices'])
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('area (sqr ft)', fontsize=20)
plt.ylabel('price (US$)', fontsize=20)
plt.scatter(df.Area, df.Prices, color='green', marker='*')
plt.plot(df.Area, reg.predict(df[['Area']]), color='blue')
```

```python
vr = [[5000]]
reg.predict(vr)
```
